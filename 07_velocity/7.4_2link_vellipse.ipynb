### Lecture 8.4 - 2 Link Velocity Example

How does the velocity of a single joint, affect the velocity of the end effector?

Let us consider a 2-link planar robot and study the **effects of joints q1 and q2** on the end-effector.

Create a 2-link model:

```python
p2 = rtb.models.DH.Planar2()
print(p2)
```

Let's assign values for the q-values of the robots. 

```python
p2.q = [0.1, 0.2]
print(p2.q)
```

We can visualize the velocity ellipsoid via the robot's plot method but by also including the vellipose option to True

```python
matplotlib.use("nbAgg")
p2.plot(p2.q, vellipse=True);
```
